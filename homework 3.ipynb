```python
from scipy import stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('hw_25000.csv', names = ['index', 'height_inches', 'weight_pounds'], delimiter=",", header=0)
data['height'] = data['height_inches'] * 2.54
data.tail(20)
```

```python
filtered_data = data.loc [(data.height >= 170) & (data.height <= 180)]
filtered_data.reset_index(inplace=True)
data20 = filtered_data.tail(20)
data20
```

```python
bins = np.linspace( 70, 180, 50 )
```

```python
plt.hist(data20['weight_pounds'], bins, alpha = 0.5)
plt.show()
```

```python
plt.hist(data['weight_pounds'], bins, alpha = 0.5)
plt.show()
```

```python
stats.ttest_ind( data20['weight_pounds'], data['weight_pounds'] )
```
